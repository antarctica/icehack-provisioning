## The Event

### The challenge

Working in teams, produce, within the time allowed, an interesting visualisation using the data-sets provided, or data-sets which are closely related.

The form this visualisation takes, be it a plot, a graph, a static output, something which moves or even something interactive, is down to you.

You can use any language, tools and processes you want - either using a virtual environment we provide, or on an environment you provide.

Staff staff will be on-hand to provide any help we can with regards to the data, including its structure and the science that underpins it.

### The prize

There are two prizes:

1. **Best visualisation** - the best visualisation produced within the time limit of this event
2. **Most promising visualisation** - the visualisation which shows the most potential if more time was available

All entries can win either prize, but no entry can win both prizes. The winning team for each prize will receive a selection of BAS products. All participants will receive a BAS goody bag.

The winners of each prize will be voted by the participants (by a show of hands) - with the event organisers having an casting vote, and who's decision is considered final.

### The aim / ethos

Though a competition, we do not intend this event to be overly competitive. Primarily we hope everyone has a positive experience, where those with more experience share their knowledge and expertise with others.

This means we would ideally like teams to be of a mixed ability.

### Global climate data - NCEP-DOE

**Location:**
`data/global-climate/ncep-doe-reanalysis-2` (see the *[Accessing data](#Accessing-data)* section for where this will be)

**Format:**
NetCDF files

**Description:** *The NCEP-DOE Reanalysis 2 project is using a state-of-the-art analysis/forecast system to perform data assimilation using past data from 1979 through the previous year. A large subset of this data is available from PSD in its original 4 times daily format and as daily averages.*

**Reference:**
> NCEP-DOE Reanalysis 2 data provided by the NOAA/OAR/ESRL PSD, Boulder, Colorado, USA, from their Web site at http://www.esrl.noaa.gov/psd/

### Global climate data - MERRA-2

**Location:**
`data/global-climate/merra-2` (see the *[Accessing data](#Accessing-data)* section for where this will be)

**Format:**
NetCDF files

**Description:** *Global monthly mean 2 m temperatures at a horizontal resolution of 0.625° longitude × 0.5° latitude. Constructed by taking the mean of data produced by the MERRA-2 assimilation scheme at 1-hourly intervals. A reanalysis assimilation scheme uses a wide variety of observations to constrain a global atmospheric forecast model to produce a spatially and temporally complete ‘best guess’ of the state of the atmosphere, even in data-sparse regions.*

**Reference:**
> Global Modelling and Assimilation Office (GMAO) (2015),MERRA-2 statD_2d_slv_Nx: 2d,Monthly,Aggregated Statistics,Single-Level,Assimilation,Single-Level Diagnostics,version 5.12.4, Greenbelt, MD, USA:Goddard Space Flight Center Distributed Active Archive Center (GSFC DAAC), Accessed 13-Mar-2016 at doi:10.5067/KVIMOMCUO83U

### Ice-core data

**Location:**
`/data/ice-cores/` (see the *[Accessing data](#Accessing-data)* section for where this will be)

**Format:**
NetCDF files

**Description:** *Geographic variability of nitrate deposition and preservation over the Greenland Ice Sheet.*

*An analysis of 96 snow pit and ice cores distributed over the Greenland ice sheet, used to determine the main drivers of variability in the preserved records of nitrate concentration. The data set provides samples from spatially distributed locations, allowing investigation of the effect of snow accumulation rate, temperature, and sublimation on nitrate concentration.*

**Reference:**
> Burkhart, J. F., R. C. Bales, J. R. McConnell, M. a Hutterli, and M. M. Frey (2009), Geographic variability of nitrate deposition and preservation over the Greenland Ice Sheet, Journal of Geophysical Research, 114(D6), 1-15, doi:10.1029/2008JD010600

### Sea-ice data

**Location:**
`/data/sea-ice/` (see the *[Accessing data](#Accessing-data)* section for where this will be)

**Format:**
NetCDF files

**Description:** *Monthly mean sea ice fractional concentration data at a horizontal resolution of 0.25° × 0.25°, held separately for the Northern (NH) and Southern (SH) hemispheres. Derived from daily (every other day before July 1987) fields of sea ice data from satellite observations using the ‘Bootstrap 2’ sea ice retrieval algorithm. Monthly means were derived as the mean of all available daily data for each calendar month.*

*The data were downloaded from http://nsidc.org/data/nsidc-0079 as polar stereographic data on a nominal 25 km × 25 km grid, then re-gridded onto a 0.25° × 0.25° grid by linear interpolation of a triangulation constructed from the data in polar stereographic space. Note that the NH data are not spatially complete: there is a circular region (whose size varies over time due to satellite changes) which is never observed due to orbit inclination: see http://nsidc.org/data/docs/daac/nsidc0079_bootstrap_seaice.gd.html#spatialcover for more details.*

**Reference:**
> Comiso, J. C. 2000, updated 2015. Bootstrap Sea Ice Concentrations from Nimbus-7 SMMR and DMSP SSM/I-SSMIS, Version 2. Boulder, Colorado USA. NASA National Snow and Ice Data Center Distributed Active Archive Center. doi: http://dx.doi.org/10.5067/J6JQLS9EJ5HU.

## The Virtual Environment
A virtual environment is available to make getting access to the data, and some tools we think might be useful for working with it, as easy as possible.

This environment is self-contained and can be used remotely, using a hosted instance, or locally using a local virtual machine. Full access is available within the environment so if you want to install your own packages and tools you can.

For *Icehack*, we will provide a number of hosted environments - though if you want to run your own locally that's no problem.

**Note: Using this environment is completely optional** - if you prefer to use your own local machine, or any other service, please do.

### What's in the virtual environment

* Based on Ubuntu 14.04 (Trusty)
* Python 2.7 with [Miniconda](http://conda.pydata.org/docs/install/quick.html) and [Iris](http://scitools.org.uk/iris/docs/latest/index.html)
* R 3.2 with [sos4R](https://cran.r-project.org/web/packages/sos4R/index.html) and a range of other packages
* Jupyter notebook with Python and R kernels

### How do I use a virtual environment

Virtual environments can be used in two ways:

1. From the browser using Jupyter notebook
2. From the command using through SSH

#### User accounts

Virtual environments contain two user accounts, one for writing code `user`, and one for managing the underlying operating system `superuer`.

**Note:** The `superuser` account does not have access to Miniconda, local pip packages or R libraries. This account should only be used for installing additional system packages or tweaking system-wide configuration files as needed.

Their logins are:

* *User* account
    * Username: `user`
    * Password: `icehack`
* *Superuser* account
    * Username: `superuser`
    * Password: `Ahh-a-seal!`

#### Jupyter notebook

Jupyter notebook is the easiest way to interact with the environment and provides support for writing Python and R code and executing it in-browser. This makes showing others really easy, as tools like Matplotlib can display directly within a notebook.

To create a new notebook:

* Go to the [Jupyter file browser](/tree)
* Choose *New* -> *Python 2* or *R*

You can also run a terminal through the browser:

* Go to the [Jupyter file browser](/tree)
* Choose *New* -> *Terminal*

A couple of things to bare in mind:

* Notebooks and terminals are executed as the `user` user account (which does not have sudo rights)
* Terminals use the `/bin/sh` shell

#### SSH

If you prefer to run code directly through the terminal, or wish to work outside the virtual machine through an IDE for example, you can connect directly to the virtual machine over the standard SSH port, i.e. `22`.

### How do I get a virtual environment

#### Hosted environments

#### Local environments

If you want to run a virtual environment locally follow these steps:

* Ensure you have either VirtualBox (free) or VMware Fusion/Workstation (paid for) installed on your local machine
* It is strongly recommended you also install [Vagrant](http://vagrantup.com) to make managing virtual machines easier

##### If you are using Vagrant

```shell
cd ~ (or somewhere suitable)
mkdir icehack
vagrant init antarctica/icehack
vagrant up
vagrant ssh-config
```

This will create a new virtual machine using the virtual machine image for the Icehack virtual environment.
As the VM image needs to be downloaded, the `vagrant up` step may take some time.

Once running, `vagrant ssh-config`, will display the IP address assigned to the virtual environment. You will need this when connecting using SSH, or accessing Jupyter from the browser.

It may be a good idea to add a hosts file entry for this to make it easier to access.

##### If you are not using Vagrant

The virtual environment is available as a OVA (a compressed virtual machine) which can be used by all major desktop virtualisation products. We specifically support VirtualBox and VMware Fusion/Workstation only.

Although designed to be interoperable, virtual machine tools (for enabling features such as shared folders) are provider specific. Therefore we provide separate OVA's for each provider. These files are available on external hard drives you can copy from - please ask [someone](#Who's-Here-to-Help) if you need one.

OVAs are located in `/environment/ovas/0.3.0/`.

Once copied to your local machine, import the OVA into your virtualisation software, configure and wait for it to start.

##### Copying data and other files

You will then need to copy the following the directories from an external hard drive (please ask [someone](#Who's-Here-to-Help) for one) to `/home/user/` in the virtual machine:
